# Q1. Implementation and Comparison of CNN Architectures

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input,Conv2D, Dense, AveragePooling2D, Flatten, MaxPooling2D
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing.image import img_to_array,load_img
from tensorflow.keras.optimizers import Adam
```

```python
#import the data
cats = []
dogs = []

import os
for fname in os.listdir('cats_and_dogs_filtered/train/cats/'):
    img = load_img(f'cats_and_dogs_filtered/train/cats/{fname}')
    cats.append(img_to_array(img))

for fname in os.listdir('cats_and_dogs_filtered/train/dogs/'):
    img = load_img(f'cats_and_dogs_filtered/train/dogs/{fname}')
    dogs.append(img_to_array(img))
```

```python
#normalistaion
train_datagen = ImageDataGenerator(rescale=1./255)
val_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)
```

```python
#splitting data
base_dir = 'cats_and_dogs_filtered'
train_dir = os.path.join(base_dir, 'train')        
validation_dir = os.path.join(base_dir, 'validation')
test_dir = os.path.join(base_dir, 'test')   

train_cats_dir = os.path.join(train_dir, 'cats')      
train_dogs_dir = os.path.join(train_dir, 'dogs')  

validation_cats_dir = os.path.join(validation_dir, 'cats')  
validation_dogs_dir = os.path.join(validation_dir, 'dogs')  

test_cats_dir = os.path.join(test_dir, 'cats')          
test_dogs_dir = os.path.join(test_dir, 'dogs')          
```

```python
#data generator for the training dataset
train_generator = train_datagen.flow_from_directory(train_dir, target_size=(28, 28),batch_size=20,class_mode='categorical')
```

```python
#data generator for the validation dataset
val_generator = val_datagen.flow_from_directory(validation_dir,target_size=(28, 28),batch_size=20,class_mode='categorical')
```

# 1) LeNet-5

```python
#create a Sequential model named "LeNet"
lenet_model = Sequential([Input(shape=(28, 28, 3)),
                          Conv2D(6, (5, 5), padding="same", activation="tanh"),
                          AveragePooling2D(pool_size=2,strides=2),
                          Conv2D(16, (5, 5), padding="valid", activation="tanh"),
                          AveragePooling2D(pool_size=2,strides=2),
                          Flatten(),
                          Dense(120, activation="sigmoid"),
                          Dense(84, activation="sigmoid"),
                          Dense(2, activation="sigmoid"),],
                          name="LeNet")

lenet_model.summary()
```

```python
#compile the model
lenet_model.compile(
    loss='binary_crossentropy',           
    optimizer=Adam(learning_rate=0.001),    
    metrics=['acc'])
```

```python
#train the model
lenet_hist = lenet_model.fit(train_generator,
                             epochs=20,
                             validation_data=val_generator)
```

# 2) AlexNet

```python
train_generator = train_datagen.flow_from_directory(train_dir,target_size=(227, 227),batch_size=20,class_mode='categorical' )
```

```python
val_generator = val_datagen.flow_from_directory(validation_dir,target_size=(227, 227),batch_size=20,class_mode='categorical',)
```

```python
#create a Sequential model named "AlexNet"
alexnet_model = Sequential([
    Input(shape=(227, 227, 3)),
    Conv2D(96, (11, 11), strides=(4, 4), padding="valid", activation="tanh"),
    MaxPooling2D(pool_size=(3, 3), strides=2),
    Conv2D(256, (5, 5), padding="same", activation="tanh"),
    MaxPooling2D(pool_size=(3, 3), strides=2),
    Conv2D(384, (3, 3), padding="same", activation="tanh"),
    Conv2D(384, (3, 3), padding="same", activation="tanh"),
    Conv2D(256, (3, 3), padding="same", activation="tanh"),
    MaxPooling2D(pool_size=(3, 3), strides=2),
    Flatten(),
    Dense(4096, activation="sigmoid"),
    Dense(4096, activation="sigmoid"),
    Dense(2, activation="sigmoid"),],
    name="AlexNet")
alexnet_model.summary()
```

```python
#compile the model
alexnet_model.compile(loss='binary_crossentropy', optimizer=Adam(learning_rate=0.001), metrics=['acc'])
```

```python
#train the model
alexnet_hist = alexnet_model.fit(train_generator, epochs=20, validation_data=val_generator)
```

# Q2. Transfer Learning and Model Performance Analysis

```python
from keras.applications import VGG16, ResNet50, MobileNetV2, EfficientNetB0, InceptionV3
```

```python
train_generator = train_datagen.flow_from_directory(train_dir,target_size=(224, 224),batch_size=20,class_mode='categorical')
```

```python
val_generator = val_datagen.flow_from_directory(validation_dir,target_size=(224, 224),batch_size=20,class_mode='categorical')
```

## VGG-16

```python
vgg_model = VGG16(input_shape=(224, 224, 3), weights='imagenet', include_top=False)
vgg_model.trainable = False
vgg_model.summary()
```

```python
custom_vgg_model = Sequential([
    vgg_model,               
    Flatten(),                
    Dense(2, activation='softmax')  
])
#custom VGG16-based model
custom_vgg_model.compile(loss='categorical_crossentropy',
                         optimizer=Adam(learning_rate=0.0001),
                         metrics=['acc'])

custom_vgg_model.summary()
```

```python
#train the model
vgg_history = custom_vgg_model.fit(train_generator,
                                   epochs=20,
                                   validation_data=val_generator)
```
